```python
# Create a KML of the PIXC swath using a Convex Hull

import numpy as np
from scipy.spatial import ConvexHull
import simplekml

# Extract coordinates
lon = data.longitude.values
lat = data.latitude.values

# Build Nx2 array
points = np.column_stack((lon, lat))

# Remove NaNs
points = points[np.isfinite(points).all(axis=1)]

# Remove impossible coordinates
points = points[
    (points[:,0] >= -180) &
    (points[:,0] <= 180) &
    (points[:,1] >= -90) &
    (points[:,1] <= 90)
]

# Remove duplicate coordinates
points = np.unique(points, axis=0)

# Need at least three points
if len(points) < 3:
    raise ValueError("Not enough valid points to build a Convex Hull.")

# Compute Convex Hull
hull = ConvexHull(points, qhull_options="QJ")

# Hull vertices are already ordered around the polygon
boundary = points[hull.vertices]

# Close the polygon
boundary = np.vstack((boundary, boundary[0]))


print("Hull vertices:", len(boundary))

# Create KML
kml = simplekml.Kml()

polygon = kml.newpolygon(name="SWOT PIXC Swath")
polygon.outerboundaryis = boundary.tolist()

polygon.style.linestyle.width = 3
polygon.style.polystyle.fill = 0

kml.save("Bristol.kml")

print("Saved Bristol.kml")
```

```python
# Create a KML of the PIXC swath using a Bounding Box

import numpy as np
import simplekml

# Extract coordinates
lon = data.longitude.values.flatten()
lat = data.latitude.values.flatten()

# Build Nx2 array
points = np.column_stack((lon, lat))

# Remove NaNs
points = points[np.isfinite(points).all(axis=1)]

# Remove impossible coordinates
points = points[
    (points[:,0] >= -180) &
    (points[:,0] <= 180) &
    (points[:,1] >= -90) &
    (points[:,1] <= 90)
]

# Remove duplicate coordinates
points = np.unique(points, axis=0)


# Need enough points
if len(points) < 4:
    raise ValueError("Not enough valid points to build a swath boundary.")


# Create rectangular boundary
min_lon = points[:,0].min()
max_lon = points[:,0].max()
min_lat = points[:,1].min()
max_lat = points[:,1].max()


boundary = np.array([
    [min_lon, min_lat],
    [max_lon, min_lat],
    [max_lon, max_lat],
    [min_lon, max_lat],
    [min_lon, min_lat]
])


print("Boundary vertices:", len(boundary))
print("Longitude:", min_lon, max_lon)
print("Latitude:", min_lat, max_lat)


# Create KML
kml = simplekml.Kml()

polygon = kml.newpolygon(name="SWOT PIXC Swath")

polygon.outerboundaryis = [
    (float(x), float(y)) for x, y in boundary
]

polygon.style.linestyle.width = 3
polygon.style.polystyle.fill = 0

kml.save("Bristol.kml")

print("Saved Bristol.kml")
```

```python
import geopandas as gpd
from shapely.geometry import Polygon, Point
import matplotlib.colors as mcolors
import numpy as np
import matplotlib.pyplot as plt

# Read KML swath boundary
kml = gpd.read_file(
    r"C:\Users\pmalesza\Documents\SWOT kml\subset_bristol.kml"
)

geom = kml.geometry.iloc[0]

if geom.geom_type == "LineString":
    poly = Polygon(geom.coords)
else:
    poly = geom


# Extract PIXC coordinates
lon = data.longitude.values
lat = data.latitude.values


# Create mask for points inside KML polygon
mask = np.array([
    poly.covers(Point(lon, lat))
    for lon, lat in zip(
        data.longitude.values,
        data.latitude.values
    )
])
```

```python
from skimage.morphology import remove_small_objects, binary_closing, disk
from scipy.ndimage import binary_fill_holes

az_sub = az[mask]
rg_sub = rg[mask]
class_sub = data.classification.values[mask]

az_min, az_max = az_sub.min(), az_sub.max()
rg_min, rg_max = rg_sub.min(), rg_sub.max()

n_az = az_max - az_min + 1
n_rg = rg_max - rg_min + 1


class_grid = np.zeros((n_az, n_rg), dtype=bool)
class_grid[az_sub - az_min, rg_sub - rg_min] = (class_sub == 1)


populated = np.zeros((n_az, n_rg), dtype=bool)
populated[az_sub - az_min, rg_sub - rg_min] = True


BW = class_grid | ~populated
se = disk(2)  

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12, 8))  

axes[0, 0].imshow(BW, origin='lower', aspect='auto')
axes[0, 0].set_title('Initial Gridded Classification Index')

BW1 = remove_small_objects(BW, min_size=50, connectivity=2)   
axes[0, 1].imshow(BW1, origin='lower', aspect='auto')
axes[0, 1].set_title('Index with bwareaopen applied')

BW2 = binary_closing(BW1, se)                                 
axes[1, 0].imshow(BW2, origin='lower', aspect='auto')
axes[1, 0].set_title('Index with bwareaopen & imclose applied')

BW3 = binary_fill_holes(BW2)                                  
axes[1, 1].imshow(BW3, origin='lower', aspect='auto')
axes[1, 1].set_title('Index with bwareaopen, imclose & imfill applied')

plt.tight_layout()
plt.show()
```

```python
"""
Python conversion of MATLAB SWOT L2 HR PIXC gridding + binary mask cleaning script.

Requires: netCDF4, numpy, matplotlib, scikit-image, scipy
    pip install netCDF4 numpy matplotlib scikit-image scipy
"""

import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
from skimage.morphology import remove_small_objects, binary_closing, disk
from scipy.ndimage import binary_fill_holes

# %% --- Step 0: Read SWOT file ---
file = ( r"\Users\pmalesza\Documents\Python Codes\SWOT_L2_HR_PIXC_052_475_246R_20260706T065938_20260706T065949_PID0_01.nc")

with Dataset(file, 'r') as nc:
    pixc = nc.groups['pixel_cloud']
    tvp = nc.groups['tvp']

    swot = {
        'lat': pixc.variables['latitude'][:],
        'lon': pixc.variables['longitude'][:],
        'lat_sat': tvp.variables['latitude'][:],
        'lon_sat': tvp.variables['longitude'][:],
        # 'height': pixc.variables['height'][:],
        'classification': pixc.variables['classification'][:],
    }

    # Read the indices from the pixel_cloud group
    az_idx = pixc.variables['azimuth_index'][:]
    rg_idx = pixc.variables['range_index'][:]

# Indices are 0-based in the product -> Python is already 0-based, no shift needed
az_idx = np.asarray(az_idx, dtype=int)
rg_idx = np.asarray(rg_idx, dtype=int)

# Array dimensions: use the product's interferogram size if known,
# otherwise size to the max index present (+1 because 0-based)
n_az = az_idx.max() + 1
n_rg = rg_idx.max() + 1

# Place each point directly into the radar-geometry array
lat_grid = np.full((n_az, n_rg), np.nan)
lon_grid = np.full((n_az, n_rg), np.nan)

lat_grid[az_idx, rg_idx] = swot['lat']
lon_grid[az_idx, rg_idx] = swot['lon']

# Boolean water mask: True where classification == 1, False elsewhere
# (unpopulated grid cells are False here; see note below re MATLAB NaN behaviour)
class_grid = np.zeros((n_az, n_rg), dtype=bool)
class_grid[az_idx, rg_idx] = (np.asarray(swot['classification']) == 1)

populated = np.zeros((n_az, n_rg), dtype=bool)
populated[az_idx, rg_idx] = True
BW = class_grid | ~populated   # replicate MATLAB's NaN-as-foreground
se = disk(2)  # strel('disk', 2)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(12, 8))  # linkaxes

axes[0, 0].imshow(BW, origin='lower', aspect='auto')
axes[0, 0].set_title('Initial Gridded Classification Index - '
                     'pixel_cloud/classification == 1')

BW1 = remove_small_objects(BW, min_size=50,connectivity=2)      # bwareaopen: remove specks < 50 px
axes[0, 1].imshow(BW1, origin='lower', aspect='auto')
axes[0, 1].set_title('Index with bwareaopen applied')

BW2 = binary_closing(BW1, se)                     # imclose: bridge small gaps
axes[1, 0].imshow(BW2, origin='lower', aspect='auto')
axes[1, 0].set_title('Index with bwareaopen & imclose applied')

BW3 = binary_fill_holes(BW2)                      # imfill(BW, 'holes')
axes[1, 1].imshow(BW3, origin='lower', aspect='auto')
axes[1, 1].set_title('Index with bwareaopen, imclose & imfill applied')

plt.tight_layout()
plt.show()

# bwareaopen(~BW, 500): kills any false region < 500 px, border-touching or not
#BW_i = remove_small_objects(~BW, min_size=500)

plt.figure(figsize=(8, 8))
plt.imshow(BW.astype(int) - BW1.astype(int), origin='lower', cmap='Blues')
plt.show()
```
